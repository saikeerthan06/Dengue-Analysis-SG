```python
# Clean and standardize column names for the 2019 data
csv_2019.columns = csv_2019.iloc[0]
csv_2019.columns = csv_2019.columns.str.strip()
csv_2019 = csv_2019.iloc[1:]  # remove the first row which is now the column names

# Clean and standardize column names for the 2020 data
csv_2020.columns = csv_2020.iloc[0]
csv_2020.columns = csv_2020.columns.str.strip()
csv_2020 = csv_2020.iloc[1:]  # remove the first row which is now the column names

# Clean and standardize column names for the 2021 data
csv_2021.columns = csv_2021.iloc[0]
csv_2021.columns = csv_2021.columns.str.strip()
csv_2021 = csv_2021.iloc[1:]  # remove the first row which is now the column names

# Clean and standardize column names for the 2022 data
csv_2022.columns = csv_2022.iloc[0]
csv_2022.columns = csv_2022.columns.str.strip()
csv_2022 = csv_2022.iloc[1:]  # remove the first row which is now the column names

# Rename columns to make them unique
csv_2019.columns = [f"2019_{col}" for col in csv_2019.columns]
csv_2020.columns = [f"2020_{col}" for col in csv_2020.columns]
csv_2021.columns = [f"2021_{col}" for col in csv_2021.columns]
csv_2022.columns = [f"2022_{col}" for col in csv_2022.columns]

# Merge the two datasets
df_merged1 = pd.concat([csv_2019, csv_2020], axis=1)
df_merged2 = pd.concat([df_merged1, csv_2021], axis=1)
df_merged3 = pd.concat([df_merged2, csv_2022], axis=1)

# Display the merged dataframe
df_merged3
```

```python
df_merged3.to_csv('hospital_final.csv')
```
